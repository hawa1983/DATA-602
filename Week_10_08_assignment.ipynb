### Convert to the appropiate data type

Below are various data cleaning and transformation steps, which are crucial for several reasons:

* ***Date Conversion:*** We will Convert the date columns from object to datetime format which enables the use of datetime operations, which are optimized for time series analyses, like time-based grouping, sorting, and filtering.
* ***Time Formatting:*** We will tranform the time data to ensure consistency, which is important for time-based comparisons and visualizations.
* ***Categorical Conversion:*** Here, we Change the appropriate columns to 'category' type which will reduce memory usage and speeds up operations like sorting and grouping. This is also essential for model training.
* ***Text Standardization:*** Converting text to upper case ensures uniformity, crucial for consistent string comparisons and aggregation.
* ***Data Integrity:*** Padding numeric IDs with leading zeros ensures that all IDs have a consistent format, important for accurate record matching and joins.

```
from datetime import datetime

# Convert date columns to datetime objects
crime_data['Date Rptd'] = pd.to_datetime(crime_data['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')
crime_data['DATE OCC'] = pd.to_datetime(crime_data['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# Function to convert military time integer to time object
def convert_military_time(military_time):
    # Format the integer as a string with leading zeros, to ensure it has a length of 4
    time_str = f'{military_time:04d}'
    # Insert the colon to separate hours and minutes
    time_str = time_str[:2] + ':' + time_str[2:]
    # Parse the string as a time object
    return datetime.strptime(time_str, '%H:%M')

# Apply the conversion function to the 'TIME OCC' column
crime_data_Cleaned = crime_data.copy()
crime_data_Cleaned['TIME OCC'] = crime_data_Cleaned['TIME OCC'].apply(convert_military_time)
# Filter out records from the year 2024

# Standardize Text: Standardize the string columns by converting to upper case.
# List of column names to convert to upper case
columns_to_upper = ['Crm Cd Desc', 'AREA NAME', 'Premis Desc', 'Weapon Desc', 'Crm Cd Desc',
                    'LOCATION', 'Cross Street', 'Weapon Desc', 'Status Desc']

# Convert each listed column to categorical
for column in columns_to_upper:
    crime_data_Cleaned[column] = crime_data_Cleaned[column].str.upper().astype('str')

# Convert `Rpt Dist No` to categorical
# First convert numbers to 4-digit strings with padding
crime_data_Cleaned['Rpt Dist No'] = crime_data_Cleaned['Rpt Dist No'].apply(lambda x: f'{x:04}')

# List of column names to convert to categorical type
columns_to_categorical = ['AREA', 'Rpt Dist No', 'Part 1-2', 'Crm Cd', 'Vict Sex',
                      'Vict Descent', 'Premis Cd', 'Weapon Used Cd', 'Status',
                      'Status Desc']

# Convert each listed column to categorical
for column in columns_to_categorical:
    crime_data_Cleaned[column] = crime_data_Cleaned[column].astype('category')

# Verify the conversion
print(crime_data_Cleaned.dtypes)
```

### Handle Numerical Data: Clean 'Vict Age' column for incorrect values.
Assuming age can't be negative


```
crime_data_Cleaned['Vict Age'] = crime_data_Cleaned['Vict Age'].clip(lower=0)
```

### Location Data: Validate latitude and longitude values.
Assuming Longitude and Latitude cannot be zero

```
crime_data_Cleaned = crime_data_Cleaned[(crime_data_Cleaned['LAT'] != 0) | (crime_data_Cleaned['LON'] != 0)]
```

### Creating New Features: Create new columns as needed.

Below, we will create new features from the existing ones to improve the performance. We could use this for machine learning models or to provide more insightful analysis. Here’s the significance of why we will do this:

* ***Creating Time-Related Features:*** We will extract 'Year', 'Month', 'Day', and 'Hour' from date-time fields to help us to uncover patterns and trends over these time intervals. It will also allow us to perform time-series analysis to help understand the cyclical nature of data.
* ***Categorical Conversion:*** Converting the 'Year' to a categorical data type will help in situations where the year is used as a category, such as in classification models or grouped visualizations.
* ***Clustering Latitude and Longitude:*** Applying K-Means clustering to geographical coordinates (latitude and longitude) identifies clusters of crime locations, which can be useful for identifying crime hotspots or for use in geospatial analysis.
* ***Removing Unwanted Data:*** Filtering out data for the year 2024 is necessary because this data is incomplete.

```
from sklearn.cluster import KMeans

crime_data_Cleaned['Days Between'] = (crime_data_Cleaned['Date Rptd'] - crime_data_Cleaned['DATE OCC']).dt.days

# Create new columns if necessary
crime_data_Cleaned['Year'] = crime_data_Cleaned['DATE OCC'].dt.year
crime_data_Cleaned['Month'] = crime_data_Cleaned['DATE OCC'].dt.month
crime_data_Cleaned['Day'] = crime_data_Cleaned['DATE OCC'].dt.day
# crime_data_Cleaned['Hour of Day'] = crime_data_Cleaned['TIME OCC'].astype(str).str.zfill(4).str[:2].astype(int) + crime_data_Cleaned['TIME OCC'].astype(str).str.zfill(4).str[2:].astype(int) / 60
crime_data_Cleaned['Hour'] = crime_data_Cleaned['TIME OCC'].dt.hour

crime_data_Cleaned['Day of Week'] = pd.to_datetime(crime_data_Cleaned['DATE OCC']).dt.day_name()

# Perform K-Means clustering on the LAT/LON data
kmeans = KMeans(n_clusters=50)  # 50 is an arbitrary choice for the number of clusters
crime_data_Cleaned[['LAT', 'LON']] = crime_data_Cleaned[['LAT', 'LON']].fillna(0)  # Replace NaN with 0 or use another imputation strategy
crime_data_Cleaned['Cluster'] = kmeans.fit_predict(crime_data_Cleaned[['LAT', 'LON']])

crime_data_Cleaned = crime_data_Cleaned[crime_data_Cleaned['Year'] != 2024]
crime_data_Cleaned.head()
```

## Part 3: Differences Between Matplotlib and Seaborn
  * **Abstraction Level:** Matplotlib offers a more granular level of control which is great for custom visualizations, while Seaborn provides a higher-level interface for creating more complex charts with less code.
  * **Default Aesthetics:** Seaborn has better default aesthetics and built-in themes that are designed to be visually pleasing and are often more statistically informative.
  * **Statistical Plots:** Seaborn excels at creating complex statistical plots and automatically includes elements like error bars.
  * **Integration with Pandas:** Seaborn works seamlessly with Pandas DataFrames, making it easier to plot data directly from dataframes.
  * **Faceting:** Seaborn's ***`FacetGrid`*** is a powerful feature for creating multiple axes arrayed in a grid that can represent different subsets of the data. While matplotlib can also create subplots, Seaborn's implementation is more integrated with data frames and more straightforward to use when working with categorical data.
