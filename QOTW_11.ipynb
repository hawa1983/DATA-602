## 1. How do you plot a histogram in Seaborn?
To plot a histogram in Seaborn, you can use the histplot function. This function allows you to create a histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
data = sns.load_dataset('tips')

# Create a histogram of the total bill
sns.histplot(data['total_bill'])
plt.show()
```

## 2. Plot a histogram with NAs dropped.
For a dataset that contains missing values (NaNs), you need to plot a histogram without these, you can use the dropna() method to exclude them.

```python
# Create a histogram of the total bill, dropping NAs
sns.histplot(data['total_bill'].dropna())
plt.show()
```

## 3. How do you set the color for a histogram?
You can set the color of the histogram by using the color parameter in histplot.

```python
# Create a histogram of the total bill with a specific color
sns.histplot(data['total_bill'], color='red')
plt.show()
```

## 4. What type of plot would allow you to compare two continuous features? Give an example of code.
To compare two continuous features, you can use a scatter plot. This type of plot lets you visualize how the two variables relate to each other.

```python
# Load a dataset
data = sns.load_dataset('tips')
# Create a scatter plot comparing total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.show()
```

## 5. Give an example of a correlation plot.
A correlation plot is useful for visualizing the correlation coefficients between variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = sns.load_dataset('tips')

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr = numeric_data.corr()

# Generate a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

6. Change the figure size of your plot(s).
To change the figure size, you can use matplotlib.pyplot's figure function before creating your plot.

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create a histogram of the total bill
sns.histplot(data['total_bill'])
plt.show()
```
